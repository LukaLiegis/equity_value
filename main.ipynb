```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error 
```

```python
data = pd.read_csv('data/actg.csv', on_bad_lines="skip", decimal=",", sep=";", index_col=0)
```

```python
data
```

```python
X = np.array(data.mc).reshape(-1, 1)
```

```python
y = data.drop("mc", axis=1)
```

```python
# creating train and test sets 
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2) 
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
data = scaler.transform(data)
```

```python
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# Initialize and fit the linear regression model
model_linear = LinearRegression()
model_linear.fit(X_train, y_train)
```

```python
# making predictions 
predictions = model_linear.predict(X_test) 
```

```python
# model evaluation 
print( 'mean_squared_error : ', mean_squared_error(y_test, predictions)) 
print( 'mean_absolute_error : ', mean_absolute_error(y_test, predictions)) 
```

```python
ridge_model = Ridge()
```

```python
ridge_model.fit(X_train, y_train)
```

```python
ridge_predictions = ridge_model.predict(X_test)
```

```python
# model evaluation 
print( 'mean_squared_error : ', mean_squared_error(y_test, ridge_predictions)) 
print( 'mean_absolute_error : ', mean_absolute_error(y_test, ridge_predictions)) 
```
